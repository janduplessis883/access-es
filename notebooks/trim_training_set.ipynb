# Trimming **Training Set**
Trimming the training set to make it possible to add the current FY data to the end of the training set.  
training till 31 March 2025 the 1 April 25 to 31 March 2026 to become part of the training set.

```python
import pandas as pd
```

```python
# Training Set 
dfs = []
for i in range(1,11):
    df = pd.read_csv(f'/Users/janduplessis/Library/CloudStorage/OneDrive-NHS/python-data/ACCESS BI/train_{i}.csv')
    dfs.append(df)

train = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
train.shape
```

```python
train.head()
```

```python
train.columns = train.columns.str.lower().str.replace(' ', '_').str.strip()
```

```python
train.head(2)
```

```python
train['rota_type'].value_counts()
```

```python
selected_rota_types = ['Stanhope Mews GP','Stanhope Mews Nurse','Stanhope Mews Nurse Practitioner','Stanhope Mews Registrar',
                       'Session','Stanhope Mews Staff','Med Student Session','Urgent Clinics','Tutorial Session', 'GP Registrar Session',
                       'MCMW Frailty MDT','GP Session','Nurse Clinic']
```

```python
train_final = train[train['rota_type'].isin(selected_rota_types)].copy()
```

```python
train_final.shape
```

```python
train_final.to_csv('/Users/janduplessis/Library/CloudStorage/OneDrive-NHS/python-data/ACCESS BI/FILTERED_TRAIN_SET.csv', index=False)
```

```python
train_final['appointment_date'] = pd.to_datetime(train_final['appointment_date'], format='%d-%b-%y')
```

```python
trimmed_train = train_final[train_final['appointment_date'] < pd.Timestamp('2025-04-01')]
```
